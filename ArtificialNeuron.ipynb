```python
inputs = [0.5, 0.3, 0.2]
weights = [0.4, 0.7, 0.2]
```

```python
import math
```

```python
def activate(inputs, weights):
  h = sum(x * w for x, w in zip(inputs, weights))
  return sigmoid(h)

def sigmoid(h):
  return 1 / (1 + math.pow((math.e), -h))
```

```python
output = activate(inputs, weights)
output
```
